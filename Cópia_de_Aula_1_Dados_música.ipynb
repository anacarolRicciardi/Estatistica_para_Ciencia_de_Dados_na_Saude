```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configuração da figura para acomodar todos os boxplots
plt.figure(figsize=(12, 8))

# Criação dos boxplots para todas as colunas em 'Quanti'
sns.boxplot(data=Quanti, palette='Reds')

# Customização do eixo x
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos das colunas para melhor legibilidade
plt.ylabel('Valores')

# Título do gráfico
plt.title('Boxplots para Variáveis Quantitativas')

# Exibição do gráfico
plt.show()
```

```python
# Define o número de colunas e linhas para os subplots
num_columns = 5
num_rows = len(Quanti.columns) // num_columns + (len(Quanti.columns) % num_columns > 0)

# Configura o tamanho da figura
fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, num_rows * 3))

# Plota cada histograma
for i, column in enumerate(Quanti.columns):
    row = i // num_columns
    col = i % num_columns
    axes[row, col].hist(Quanti[column], bins=20)
    axes[row, col].set_title(f'Histograma de {column}')
    axes[row, col].set_xlabel(column)
    axes[row, col].set_ylabel('Frequência')

# Ajusta o layout para não sobrepor os subplots
plt.tight_layout()
plt.show()
```

```python

# Define o número de colunas e linhas para os subplots
num_columns = 5
num_rows = len(Quanti.columns) // num_columns + (len(Quanti.columns) % num_columns > 0)

# Configura o tamanho da figura
fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, num_rows * 3))

# Plota cada densidade alisada
for i, column in enumerate(Quanti.columns):
    row = i // num_columns
    col = i % num_columns
    sns.kdeplot(Quanti[column], ax=axes[row, col], fill=True, color='blue', alpha=0.6) #  sns.kdeplot mostra as densidades
    axes[row, col].set_title(f'Densidade Alisada de {column}')
    axes[row, col].set_xlabel(column)
    axes[row, col].set_ylabel('Densidade')

# Ajusta o layout para não sobrepor os subplots
plt.tight_layout()
plt.show()
```

### Gráficos de mosaico

```python
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt


# Cria a figura e os eixos
fig, ax = plt.subplots(figsize=(15, 10))

# Cria o gráfico de mosaico
mosaic(df, ['Tom', 'Modo'], title='Gráfico de Mosaico: Tom por Modo',
       properties=dict(boxdict={'facecolor': 'lightblue', 'edgecolor': 'black', 'linewidth': 2}),
       labelizer=lambda k: f'{k[0]}\n{k[1]}',
       gap=0.01, ax=ax)

# Ajusta os rótulos e título
ax.set_title('Gráfico de Mosaico: Tom por Modo', fontsize=16)
ax.set_xlabel('Tom', fontsize=14)
ax.set_ylabel('Modo', fontsize=14)

# Ajusta o layout e exibe o gráfico
plt.show()
```
